```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.drop(["Unnamed: 32", "id"], inplace=True, axis=1)
```

```python
import seaborn as sns
```

```python
sns.countplot(data["diagnosis"])
print(data.diagnosis.value_counts())
```

```python
diagnosis_ = []
for i in data['diagnosis']:
    if i == 'M':
        diagnosis_.append('1')
    elif i == 'B':
        diagnosis_.append('0')
data['diagnosis_'] = diagnosis_
```

```python
data.head()
```

```python
data.drop('diagnosis', axis=1, inplace=True)
data.head()
```

```python
from sklearn.model_selection import train_test_split
x = data.drop(["diagnosis_"], axis=1)
y = data.diagnosis_

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
scaled_x = scaler.transform(x)

train_x, test_x, train_y, test_y = train_test_split(scaled_x, y, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(4)
knn.fit(train_x, train_y)
y_pred = knn.predict(test_x)
print("Test set score: {:.2f}".format(knn.score(test_x, test_y)))
print("Training set score: {:.2f}".format(knn.score(train_x, train_y)))
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(knn, x, y, cv=7)
print(cv_scores)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='lbfgs', max_iter=100, penalty='none')
logreg.fit(train_x, train_y)
y_pred = logreg.predict(test_x)
print("Test set score: {:.2f}".format(logreg.score(test_x,test_y)))
print("Training set score: {:.2f}".format(logreg.score(train_x, train_y)))
```

```python
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(test_y,y_pred).ravel()
precision = tp/(tp + fp)
sensitivity = tp/(tp + fn)
specificity = tn/(tn + fn)
false_positive_rate = 1 - specificity
print(precision, sensitivity, specificity, false_positive_rate)
```

```python
# logistic regression model
# With penalty
logreg_wp = LogisticRegression(solver='lbfgs', max_iter=100, penalty='l2')
logreg_wp.fit(train_x, train_y)
y_pred = logreg_wp.predict(test_x)
```

```python
print("Test set score: {:.2f}".format(logreg_wp.score(test_x,test_y)))
print("Training set score: {:.2f}".format(logreg_wp.score(train_x, train_y)))
```
