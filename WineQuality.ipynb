```python
# Make quality good/bad
# Build a model
```

```python
quality_ = []
for i in wine_data_all['quality']:
    if 1 <= i <= 5:
        quality_.append('1')
    elif 6 <= i <= 10:
        quality_.append('2')
wine_data_all['quality_'] = quality_
```

```python
wine_data_all.columns
wine_data_all.head()
```

```python
wine_data_all.drop('quality', axis=1, inplace=True)
```

```python
# dividing data
from sklearn.model_selection import train_test_split
x = wine_data_all.drop(["quality_"], axis=1)
y = wine_data_all.quality_

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
scaled_x = scaler.transform(x)

train_x, test_x, train_y, test_y = train_test_split(scaled_x, y, random_state = 42)
```

```python
# Train model using different number of neighbors
from sklearn.neighbors import KNeighborsClassifier
for i in range(15):
    knn = KNeighborsClassifier(n_neighbors=i + 1)
    knn.fit(train_x, train_y)
    y_pred = knn.predict(test_x)
    print("Scores for i =", i + 1)
    print("Test set score: {:.2f}".format(knn.score(test_x, test_y)))
    print("Training set score: {:.2f}".format(knn.score(train_x, train_y)))
```

```python
knn = KNeighborsClassifier(3)
knn.fit(train_x, train_y)
y_pred = knn.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
#print(y_pred)
#print(test_y)
print(confusion_matrix(test_y,y_pred).ravel())
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(knn, x, y, cv=7)
print(cv_scores)
```

```python
# logistic regression model
# Without penalty
train_x_logreg, test_x_logreg, train_y_logreg, test_y_logreg = train_test_split(scaled_x, y, random_state  = 365)
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='lbfgs', max_iter=100, penalty='none')
logreg.fit(train_x_logreg, train_y_logreg)
y_pred = logreg.predict(test_x)
```

```python
print("Test set score: {:.2f}".format(logreg.score(test_x_logreg,test_y_logreg)))
print("Training set score: {:.2f}".format(logreg.score(train_x_logreg, train_y_logreg)))
```

```python
cv_scores = cross_val_score(logreg, scaled_x, y, cv=7)
print(cv_scores)
```

```python
tn, fp, fn, tp = confusion_matrix(test_y_logreg,y_pred).ravel()
precision = tp/(tp + fp)
sensitivity = tp/(tp + fn)
specificity = tn/(tn + fn)
false_positive_rate = 1 - specificity
print(precision, sensitivity, specificity, false_positive_rate)
```

```python
# logistic regression model
# With penalty
train_x_logreg, test_x_logreg, train_y_logreg, test_y_logreg = train_test_split(scaled_x, y, random_state  = 365)
from sklearn.linear_model import LogisticRegression
logreg_wp = LogisticRegression(solver='lbfgs', max_iter=100, penalty='l2')
logreg_wp.fit(train_x_logreg, train_y_logreg)
y_pred = logreg_wp.predict(test_x_logreg)
```

```python
print("Test set score: {:.2f}".format(logreg_wp.score(test_x_logreg,test_y_logreg)))
print("Training set score: {:.2f}".format(logreg_wp.score(train_x_logreg, train_y_logreg)))
```

```python
tn, fp, fn, tp = confusion_matrix(test_y_logreg,y_pred).ravel()
precision = tp/(tp + fp)
sensitivity = tp/(tp + fn)
specificity = tn/(tn + fn)
false_positive_rate = 1 - specificity
print(precision, sensitivity, specificity, false_positive_rate)
```
